## Problem 1

Read in the file *airports.csv* as a pandas dataframe. Convert the dataframe into a geopandas dataframe.

Find the distance between the following airports:
1. Halifax / CFB Shearwater Heliport (Halifax, Canada) to Murtala Muhammed International Airport (Lagos, Nigeria)
2. Don Mueang International Airport (Bangkok, Thailand) and Beijing Capital International Airport (Beijing, China)
3. Salt Lake City International Airport (Salt Lake City, USA) and Auckland International Airport (Auckland, New Zealand)

```python
df = pd.read_csv("airports.csv").set_index("Airport ID")
df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
df['Coordinates'] = df['Coordinates'].apply(Point)
gdf = gpd.GeoDataFrame(df, geometry='Coordinates')
```

```python
halifax = gdf[gdf.Name == 'Halifax / CFB Shearwater Heliport'].reset_index().Coordinates
murtala = gdf[gdf.Name == 'Murtala Muhammed International Airport'].reset_index().Coordinates
don_mueang = gdf[gdf.Name == 'Don Mueang International Airport'].reset_index().Coordinates
beijing = gdf[gdf.Name == 'Beijing Capital International Airport'].reset_index().Coordinates
slc = gdf[gdf.Name == 'Salt Lake City International Airport'].reset_index().Coordinates
auckland = gdf[gdf.Name == 'Auckland International Airport'].reset_index().Coordinates
d1 = halifax.distance(murtala)
d2 = don_mueang.distance(beijing)
d3 = slc.distance(auckland)

print("Halifax to Murtala:", float(d1))
print("Don Mueang to Beijing:", float(d2))
print("SLC to Auckland:", float(d3))
```

## Problem 2
Read in the built-in GeoDataFrame *naturalearth_lowres*.
Create a GeoDataFrame that only contains information about the southern hemisphere.
Use this data to find the country with the smallest and largest area in the southern hemisphere.
Dissolve this GeoDataFrame to find the continent with the largest and smallest area in the southern hemisphere.

```python
world = gpd.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
south = world.cx[:, :0]
southern_continent = south.dissolve(by='continent', aggfunc='sum')

smallest_idx = south.area.idxmin()
largest_idx = south.area.idxmax()
smallest_country = world.name.loc[smallest_idx]
largest_country = world.name.loc[largest_idx]

smallest_continent = southern_continent.area.idxmin()
largest_continent = southern_continent.area.idxmax()

print("smallest southern country:", smallest_country)
print("largest southern country:", largest_country)
print("smallest southern continent:", smallest_continent)
print("largest southern continent:", largest_continent)
```

## Problem 3

Using the built-in GeoDataFrame \li{naturalearth_lowres} and GeoPlots, create population density plots for Asia and South America.
Use Mercator maps.

```python
world["pop_density"] = world["pop_est"] / world.area
world.to_crs({'init': 'epsg:3395'})

asia = world[world.continent=='Asia']
asia.plot(column='pop_density', cmap='OrRd', legend=True)
plt.title("Population Density map for Asia")
plt.show()

south_america = world[world.continent=="South America"]
south_america.plot(column='pop_density', cmap='OrRd', legend=True)
plt.title("Population Density map for South America")
plt.show()
```
