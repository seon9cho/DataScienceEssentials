## Problem 1
The file `DJIA.csv` contains daily closing values of the Dow Jones Industrial Average from 2006â€“2016.
Read the data into a `Series` or `DataFrame` with a `DatetimeIndex` as the index.
Drop rows with missing values, cast the `"VALUES"` column to floats, then plot the data.

```python
from datetime import datetime

djia = pd.read_csv("DJIA.csv", parse_dates=[0], index_col=0)

djia = djia.drop(djia[djia.VALUE=='.'].index)
djia.VALUE = djia.VALUE.astype(float)
djia.plot(lw=.5)
plt.show()
```

## Problem 2

The file `paychecks.csv` contains values of an hourly employee's last 93 paychecks.
He started working March 13, 2008.
This company hands out paychecks on the first and third Fridays of the month.
Read in the data, using `pd.date_range()` to generate the DatetimeIndex.
Plot the data.

```python
pc = pd.read_csv("paychecks.csv")
dr1 = pd.date_range(start='3/13/2008', periods=len(pc)/2, freq="WOM-3FRI")
dr2 = pd.date_range(start='3/13/2008', periods=len(pc)/2, freq="WOM-1FRI")
dates = dr1.union(dr2)
pc = pc.set_index(dates)
pc.plot()
plt.show()
```

## Problem 3

The file `finances.csv` contains a list of simulated quarterly earnings and expense totals from a fictional company.
Load the data into a `Series` or `DataFrame` with a `PeriodIndex` with a quarterly frequency.
Assume the fiscal year starts at the beginning of September and that the data begins in September 1978.
Plot the data.

```python
finances = pd.read_csv("finances.csv")
p = pd.period_range(start="9/1/1978", periods=len(finances), freq="Q-SEP")
finances = finances.set_index(p)
finances.plot()
plt.show()
```

## Problem 4

The file `website_traffic.csv` contains records for different visits to a fictitious website.
Read in the data, calculate the duration of each visit (in seconds), and convert the index to a `DatetimeIndex`.
Use downsampling to calculate the average visit duration by minute, and the average visit duration by hour.
Plot both results on the same graph.

```python
wt = pd.read_csv("website_traffic.csv")
index = pd.to_datetime(wt.ENTER)
duration = pd.to_datetime(wt.LEAVE) - pd.to_datetime(wt.ENTER)
dur = pd.Series([d.seconds for d in duration])
dur.index = index
hour = dur.resample("H").mean()
minute = dur.resample("T").mean()
minute.plot(lw=.4, label="Average visit per minute")
hour.plot(label="Average visit per hour")
plt.legend()
plt.ylabel("duration (min)")
plt.xlabel("Time")
plt.title("Average visit duration")
plt.show()
```

## Problem 5
Compute the following information about the DJIA dataset from Problem 1.
- The single day with the largest gain.
- The single day with the largest loss.
- The month with the largest gain.
- The month with the largest loss.

For the monthly statistics, define the gain (or loss) to be the difference between the DJIA on the last and first days of the month.

```python
diff = djia.diff()
max_gain = diff.max().VALUE
max_loss = diff.min().VALUE
print("The single day with the largest gain")
diff[diff.VALUE==max_gain]
```

```python
print("The single day with the largest loss")
diff[diff.VALUE==max_loss]
```

```python
month_diff = pd.DataFrame(columns=["date", "value"])
for year in range(2006, 2017):
    for month in range(1, 13):
        date = str(year) + "-" + str(month)
        try:
            val = djia[date].iloc[-1] - djia[date].iloc[0]
            month_diff = month_diff.append({"date":date, "value":val.VALUE}, ignore_index=True)
        except KeyError:
            continue
month_diff = month_diff.set_index(month_diff.date).drop(columns=['date'])
max_gain = month_diff.max().value
max_loss = month_diff.min().value
print("The month with the largest gain")
month_diff[month_diff.value==max_gain]
```

```python
print("The month with the largest loss")
month_diff[month_diff.value==max_loss]
```

## Problem 6

Plot the following from the DJIA dataset with a window or span of 30, 120, and 365.
- The original data points.
- Rolling average.
- Exponential average.
- Minimum rolling values.
- Maximum rolling values.

Describe how varying the length of the window changes the approximation to the data.

```python
for n in [30, 120, 365]:
    fig = plt.figure(figsize=(15,10))
    ax1 = plt.subplot(221)
    djia.plot(color="gray", lw=.5, ax=ax1)
    djia.rolling(window=n).mean().plot(color='r', ax=ax1)
    plt.title("Rolling Average")
    ax1.legend(["Actual", "Rolling"])
    
    ax2 = plt.subplot(222)
    djia.plot(color="gray", lw=.5, ax=ax2)
    djia.ewm(span=n).mean().plot(color='g', ax=ax2)
    plt.title("Exponential Average")
    ax2.legend(["Actual", "EWMA"])
    
    ax3 = plt.subplot(223)
    djia.plot(color="gray", lw=.5, ax=ax3)
    djia.rolling(window=n).min().plot(color='b', ax=ax3)
    plt.title("Minimum rolling values")
    ax3.legend(["Actual", "Minimum"])
    
    ax4 = plt.subplot(224)
    djia.plot(color="gray", lw=.5, ax=ax4)
    djia.rolling(window=n).max().plot(color='c', ax=ax4)
    plt.title("Maximum rolling values")
    ax4.legend(["Actual", "Maximum"])
    plt.suptitle("Window/Span = " + str(n))
    plt.tight_layout()
    plt.show()
```

The varying lengths of the window eliminates noise and smoothes out the figure.
