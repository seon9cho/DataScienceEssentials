```python
accidents = pd.read_pickle("final_accidents2.pickle")
drivers = pd.read_pickle("final_drivers.pickle")
drivers = drivers[drivers.AGE < 200]
df1 = pd.merge(accidents, drivers, on="ST_CASE")[["ST_CASE", "WEATHER", "SP", "DRINKING", "FATALS"]]
```

```python
drivers.plot(kind="hist", y=["DRINKING", "SPEEDREL"], bins=9, alpha=0.7)
plt.show()
pd.plotting.scatter_matrix(accidents[["HOUR", "DRUNK_DR", "FATALS"]])
plt.show()
pd.plotting.scatter_matrix(drivers[["DRINKING", "AGE", "SPEEDREL"]])
plt.show()
```

```python
df1 = pd.merge(accidents, drivers, on="ST_CASE")[["ST_CASE", "WEATHER", "SP", "DRINKING", "FATALS", "DRUNK_DR"]]
df1.plot(kind="scatter", x="DRUNK_DR", y="FATALS", alpha=.1, rot=30)
plt.show()
df1[df1.WEATHER<90].plot(kind="scatter", x="WEATHER", y="FATALS", alpha=.1, rot=30)
plt.show()
```

```python
accidents.plot(kind="scatter", x="LONGITUD", y="LATITUDE", alpha=0.002)
plt.show()
accidents.plot(kind="Hexbin", x="LONGITUD", y="LATITUDE", gridsize=30)
plt.show()
```

```python
from pydataset import data
road = data("road")
road
```

```python
# We leave out the outlier DC since it makes it impossible to visualize
road[road.popden < 10000].plot(kind="scatter", x="popden", y="fuel", alpha=1, rot=30)
plt.show()
road.plot(kind="scatter", x="drivers", y="deaths", alpha=0.8, rot=30)
plt.show()
road[road.popden < 10000].plot(kind="scatter", x="temp", y="popden")
plt.show()
```

1. Does higher population density indicate higher fuel consumption?

In general, it seems so, but there seem to be one trend between 0 < popden < 200 and another for popden > 200

2. Does a correlation exist between number of drivers and number of deaths in a state?

Definitely yes. It's a linear relationship

3. Does the temperature of a state in January affect its population density?

No. From the data, we can see that they are random.

```python
nycrimes = pd.read_csv("new_york_crime_clean.csv", index_col=0)
nycrimes
```

```python
nycrimes.describe()
```

```python
style_list = ['-'] * 10 + ['--'] * 5
nycrimes.plot(y=list(nycrimes), style=style_list, figsize=(5,8))
plt.legend(loc='upper center', bbox_to_anchor=(1.50, 0.5))
plt.show()
```

```python
nycrimes.plot(y=["ROBBERY", "FELONY DANGEROUS DRUGS (1)"])
plt.show()
nycrimes.plot(kind="scatter", x="ROBBERY", y="FELONY DANGEROUS DRUGS (1)", alpha=.8)
plt.show()
```

```python
nycrimes.plot(kind="box", y=list(nycrimes), vert=False, figsize=(20,10))
plt.show()
```

1. What are the most common and least common crimes?

Most common: Grand Larceny
Least common: Murder / manslaughter

2. Does a trend exist between robbery and drug felonies? If so, what is the trend?

Yes. It almost seems like a 1 to 1 correlation, and they have both been declining over the years

3. Which crimes have had the largest distributions?

Grand larceny of motor vehicle, since it seems to have the largest difference between its minimum and the maximum
